```python
import numpy as np
import math
```

```python
nim_dict = {}

def nim(N):
    if N in nim_dict.keys():
        return nim_dict[N]

    # N == 0 => nim = 0
    if N == 0:
        result = 0
    
    # N == 1 => nim = 1
    elif N == 1:
        result = 1

    # N == 2 => nim = 2
    elif N == 2:
        result = 2
    
    # N == 3 => nim = 3
    elif N == 3:
        result = 3
    
    else:
        #Median
        M = (N + 1)//2

        #Pins
        pins = [1 for k in range(N)]

        #Nim values for its options
        options = []

        #loop for 1 pin only
        for n in range(0, M):
            if n == 0:
                options.append(nim(N-1))
                options.append(nim(N-2))

            else:
                #1 pin
                left = nim(len(pins[0:n]))
                right = nim(len(pins[n+1:]))

                #2 pins
                left2 = nim(len(pins[0:n]))
                right2 = nim(len(pins[n+2:]))

                sum1 = left^right
                sum2 = left2^right2
                
                options.append(sum1)
                options.append(sum2)

        #determine the nim
        for n in range(max(options)+2):
            if n not in options:
                result = n
                nim_dict[N] = result
                return result

    nim_dict[N] = result
    return result
```
